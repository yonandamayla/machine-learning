PREDIKSI HARGA RUMAH BERDASARKAN JUMLAH KAMAR

```python
import numpy as np
```

```python
#buat data jumlah kamar
bedrooms = np.array([1,1,2,2,3,4,4,5,5,5,])
```

```python
#data harga rumah, asumsi dalam dollar
house_price = np.array([15000, 18000, 27000, 34000, 50000, 68000, 65000, 81000,85000, 90000])
```

```python
# menampilkan scatter plot dari dataset
#X = var independen (jumlah kamar)
#y = var dependen (harga rumah)
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(bedrooms, house_price)
```

```python
#melatih model dengan memanggil fungsi LinearRegression()
from sklearn.linear_model import LinearRegression

# latih model dengan Linear Regression.fit()
bedrooms = bedrooms.reshape(-1, 1)
linreg = LinearRegression()
linreg.fit(bedrooms, house_price)
```

```python
# menampilkan plot hubungan antara jumlah kamar dan harga rumah
plt.title("Bedrooms Vs House Price")
plt.xlabel("House Price")
plt.ylabel("Bedrooms")
plt.scatter(bedrooms, house_price)
plt.plot(bedrooms, linreg.predict(bedrooms))

#Model ini memiliki kompleksitas rendah dan bekerja sangat baik pada dataset yang memiliki hubungan linier
```
