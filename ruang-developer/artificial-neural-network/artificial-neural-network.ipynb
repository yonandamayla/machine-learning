# Artificial Neural Network
- Jaringan sistem saraf buatan -> menerima rangsangan dari luar - menuju otak  menghasilkan output

- Ada 3 lapisan : 
1. Input Layer : menerima inputan
2. Hidden Layer : hidden layer bisa lebih dari 1 layer (deep learning)
3. Output Layer : memberikan sebuah keputusan, klasifikasi

- Cara kerja ANN :
1. Input layer berisi inputan bisa berupa parameter beserta bobotnya
2. Hidden layer terjadi kalkulasi perkalian antara inputan dengan bobotnya, kemudian semua dijumlahkan dari semua parameter
3. Output layer adalah hasil akhir

# Contoh Kasus Penentuan Harga Rumah
Dalam memberi rumah ada parameter seperti ada berapa kamar tidur, kamar mandi, kolam renang, dll. Semisal rumah punya kolam renang biasanya harganya akan jauh lebih mahal dariapada rumah yang tidak memiliki kolam renang, maka hal ini kolam renang berbobot tinggi. 


Misal ada rumah dengan fasilitas :
2 Kamar Tidur
1 Kamar Mandi
Luas Tanah 100m persegi
2 tingkat lantai
Harga = 5.000.000.000

Jika dimodelkan menjadi ANN maka akan menjadi :
Input Layer             | Hidden Layer          | Output Layer
2       -- KT           
1       -- KM                                   outputnya 5.000.000.000
100     -- Luas Tanah       
2       -- Tingkat

Note : Jika memakai lib tf maka pembobotan biasanya akan dikasih nilai 


```python
import tensorflow as tf
import warnings
warnings.filterwarnings("ignore")
```

```python
# data berbentuk list, di dalam list ada list lagi
# urutan data : kamar tidur, kamar mandi, luas tanah, jumlah lantai
# [[], [], [], []]
# [harga, harga, harga, harga] = 50*100.000.000 = 5.000.000.000

rumahs = [[2, 2, 100, 2], [1, 1, 50, 1], [3, 3, 150, 4], [2, 2, 200, 5]]
hargas = [50, 8, 100, 40]
```

```python
model =  tf.keras.models.Sequential([
        tf.keras.layers.Dense(input_shape = [4], units = 1)
])
import warnings
warnings.filterwarnings("ignore")
model.summary()
# input_shape berguna untuk ada berapa neuron dalam input layer, karena di tensorflow akan diskip layer dari enuronnya
```

```python
model.compile(
    optimizer = 'sgd',
    metrics = ['acc'],
    loss = 'mse',
)
# optimizer = teknik evaluasi yg digunakan untuk jenis perhitungan bobot input 
# metrics = apa yg diukur dan ditampilkan setiap perulanagn/iterasi
# loss = valuenya adl fungsi untuk menghitung seberapa jauh tebakan model ml yg kita buat dengan yg sebenarnya
# loss MSE = data asli - data prediksi mesin. Jika ada 2 data, kemudian hasilnya dipangkat 2 kan agar tidak negatif kemudian ditambah dan dibagi banyak data
# loss MAE = hasil perhitungan yg negatif akan dimutlakkan sehingga minusnya akan hilang
```

```python
model.fit(
    rumahs, hargas, epochs = 100
)
import warnings
warnings.filterwarnings("ignore")
# fungsi fit untuk melatih model yg telah dibuat
# 1 epoch = 1 putaran pelatihan data
# NAN = not a number, jadi hasilnya tidak mencerminkan angka, hal ini karena data dummy dan kurang banyak data
```

```python
print(model.predict([[5,2,50,3]]))
```
