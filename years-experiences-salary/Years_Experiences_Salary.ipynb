```python
import pandas as pd
```

```python
df = pd.read_csv("Salary_Data.csv")
```

```python
#menampilkan 5 sample data
df.head()
```

```python
#menampilkan deskripsi singkat dari data
df.info()
```

```python
#proses filtering
# ":" = baris
# -1 artinya kolom terkecuali yang tidak perlu dimasukkan
# 1 artinya yang dimasukkan hanya kolom terakhir
X = df.iloc[:, :-1]
y = df.iloc[:, 1]
```

```python
#membagi data menjadi data test dan data train
#test size 1/3 artinya 100% : 1/3 = 30% datanya untuk data test, 70% sisanya untuk data train  
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=1/3, random_state =0)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#y_pred = y prediksi 
y_pred = regressor.predict(X_test)
```

```python
#untuk data train
import matplotlib.pyplot as plt
plt.scatter(X_train, y_train, color="green")
plt.plot(X_train, regressor.predict(X_train), color="red")
plt.title("Years Experience VS Salary")
plt.xlabel("Years Experience")
plt.ylabel("Salary")
plt.show()
```

```python
#untuk data test
plt.scatter(X_test, y_test, color="green")
plt.plot(X_train, regressor.predict(X_train), color="red")
plt.title("Years Experience VS Salary")
plt.xlabel("Years Experience")
plt.ylabel("Salary")
plt.show()
```

```python
#MEMBUAT PREDIKSI BARU
#prediksi angka baru untuk years experiences yang tidak ada pada data yang kita miliki, misal 2 tahun
#asumsi gaji dalam dollar
salary_pred = regressor.predict([[2]])
print("The salary for that amount of years experience is: ", salary_pred)
```
